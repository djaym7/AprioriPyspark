```python
from pyspark.mllib.fpm import FPGrowth
import pyspark
```

```python
sc = pyspark.SparkContext()
#sc.stop()
```

```python
data = sc.textFile("sample_fpgrowth.txt")
transactions = data.map(lambda line: line.strip().split(' '))
model = FPGrowth.train(transactions, minSupport=0.2, numPartitions=10)
result = model.freqItemsets().collect()
for fi in result:
    print(fi)
```

```python
# model was produced by FPGrowth.train() method
rules = sorted(model._java_model.generateAssociationRules(0.2).collect(), 
    key=lambda x: x.confidence(), reverse=True)
for rule in rules:
    # rule variable has confidence(), consequent() and antecedent() 
    # methods for individual value access.
    print (rule)
```
